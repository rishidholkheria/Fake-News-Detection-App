```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
true = pd.read_csv('True.csv')
```

```python
fake = pd.read_csv('Fake.csv')
```

```python
fake.tail()
```

```python
true['label'] = 1
```

```python
fake['label'] = 0
```

```python
true.head()
```

```python
fake.head()
```

```python
news = pd.concat([fake, true], axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1)
```

```python
news.head()
```

```python
news.tail()
```

```python
#Taking entire data and reshuffling 
news = news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'], axis=1, inplace=True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text) :
    #Convert whole text into lowercase
    text = text.lower()
    #Remove URLs
    text = re.sub(r'https?://\S+ |www\.\S+', '',text)
    # Remove HTML tags
    text = re.sub(r'<?>', '', text)
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Remove digits
    text = re.sub (r'\d', '', text)
    # Remove newline characters
    text = re.sub (r'\n', '', text)

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text'] #independant var
y = news['label'] #dependant var
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Spliting training and testing data 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_train
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
predict_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predict_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train, y_train)
```

```python
predict_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_dtc))
```

```python
#
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predict_dtc)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])
cm_display.plot()
plt.show()
```

```python
cm_lr = metrics.confusion_matrix(y_test, predict_lr)
cm_lr_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm_lr,
                                            display_labels=[False, True])
cm_lr_display.plot()
plt.show()
```
